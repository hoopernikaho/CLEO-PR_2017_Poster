```python
# find minimum threshold between n=0 and n=1 population
# fit from this point onwards to2 gaussians, with Poisson amplitude constraints
# save fitted data into g1_component.dat and g2_component.dat
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
from matplotlib.ticker import MultipleLocator
matplotlib.rcParams.update({'font.size': 18})
import thres
```

```python
# load data set containing several types of areas
dt = 2 #ns
scale = 1e3 #change from volts to mV
bins = 200
data_cont_dtleft0 = np.loadtxt('data_cont_dtleft0_dtright1300_hth_discrminated7.73E-03.dat')
```

```python
abs_area_disc = data_cont_dtleft0[:,0]
abs_area_above_th = data_cont_dtleft0[:,1]
area = data_cont_dtleft0[:,2]

np.savetxt('abs_area_disc.dat',abs_area_disc)
np.savetxt('abs_area_above_th.dat',abs_area_above_th)
np.savetxt('area.dat',area)
```

```python
plt.figure(figsize=(8,6))
plt.hist(data_cont_dtleft0[:,0]*dt*scale,bins,histtype='step',label='area (with discriminator)');
plt.hist(data_cont_dtleft0[:,1]*dt*scale,bins,histtype='step',label='area (above $V_{th}$)');
plt.hist(data_cont_dtleft0[:,2]*dt*scale,bins,histtype='step',label='area');

plt.legend()
plt.ylim(0,8000)
plt.xlim(0,20*dt*scale)

ticks = (plt.gca().get_xticks()*10**-3).astype('int')
plt.gca().set_xticklabels(ticks)


plt.xlabel('area (mV $\mu$s)')
plt.ylabel('count')

plt.savefig('area_histos_comparison.eps')
plt.show()
```

```python
# take a peek at the area distribution to be fitted
%matplotlib notebook
plt.figure(figsize=(8,6))
plt.hist(abs_area_disc,bins,range=[0,15]);
plt.ylim(0,400)
# plt.yscale('symlog')
plt.show()
```

```python
pnr_area = np.histogram(abs_area_disc*dt*scale, bins)
```

```python
# algorithm that fits and generates components
reload(thres)
result = thres.gauss_fit_poiss_ph_region(pnr_area,
                                          min_peak_sep=3*dt*scale, 
                                          threshold=1e-2, 
                                          weighted=True,
                                          plot=True)
plt.ylim(0,6000)
plt.xlim(0,15*dt*scale)
```

```python
result.best_values
```

```python
result.best_values['g2_center']/result.best_values['g1_center']
```

```python
# Prepare histogram files and save
```

```python
frequencies = pnr_area[0]
x_val = pnr_area[1]
step = np.diff(x_val)[0]
x_val = x_val[:-1] + step / 2.
```

```python
[g1, g2] = [result.eval_components(x=x_val)['g{}_'.format(k+1)] for k,_ in enumerate(result.components)]
```

```python
np.savetxt('pnr_area.dat',zip(x_val,frequencies))
np.savetxt('g1_area_component.dat',np.array(zip(x_val,g1)))
np.savetxt('g2_area_component.dat',np.array(zip(x_val,g2)))
```

```python
plt.figure(figsize=(8,6))
plt.errorbar(x_val,frequencies,
             yerr=np.sqrt(frequencies),
             linestyle='-',
             ecolor='black',
             color='black',
            label='$C(a)$'
            )
plt.plot(x_val,g1,color='blue',label='n=1')
plt.plot(x_val,g2,color='red',label='n=2')
plt.axvline(5604.84593463,color='black', label=r'$th_{01}$')
plt.axvline(15921.253917,color='black', label=r'$th_{12}$', linestyle='--')

plt.ylim(0,6000)
plt.xlim(0,15*dt*scale)

ticks = (plt.gca().get_xticks()*10**-3).astype('int')
plt.gca().set_xticklabels(ticks)
plt.xlabel('area (mV $\mu$s)')
plt.ylabel('count')

plt.legend()
plt.savefig('area_histo_with_fit.eps')
```
